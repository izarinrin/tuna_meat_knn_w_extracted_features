```python
import cv2
import os
import numpy as np

# Function to load, crop to square, resize, and remove background from an image
def load_crop_resize_remove_background(file_path, target_size):
    # Load the image with alpha channel
    image = cv2.imread(file_path, cv2.IMREAD_UNCHANGED)
    if image is None:
        return None
    
    # Crop the image to square
    height, width, _ = image.shape
    if height > width:
        start = (height - width) // 2
        cropped_image = image[start:start+width, :]
    elif width > height:
        start = (width - height) // 2
        cropped_image = image[:, start:start+height]
    else:
        cropped_image = image
    
    # Resize the image to target size
    resized_image = cv2.resize(cropped_image, (target_size, target_size))
    
    # Convert the image to grayscale
    gray = cv2.cvtColor(resized_image, cv2.COLOR_BGR2GRAY)

    # Apply Gaussian blur to reduce noise
    blurred = cv2.GaussianBlur(gray, (5, 5), 0)

    # Perform adaptive thresholding to separate foreground from background
    _, thresholded = cv2.threshold(blurred, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

    # Invert the thresholded image
    thresholded = cv2.bitwise_not(thresholded)

    # Find contours in the thresholded image
    contours, _ = cv2.findContours(thresholded, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Create a mask to store the foreground pixels
    mask = np.zeros_like(gray)

    # Draw contours on the mask
    cv2.drawContours(mask, contours, -1, (255), thickness=cv2.FILLED)

    # Bitwise AND operation to extract foreground
    foreground = cv2.bitwise_and(resized_image, resized_image, mask=mask)
    
    return foreground

# Function to crop zoom the image
def crop_zoom(image, zoom_factor):
    # Get image dimensions
    height, width, _ = image.shape
    
    # Calculate new dimensions after zoom
    new_height = int(height / zoom_factor)
    new_width = int(width / zoom_factor)
    
    # Crop zoom the image
    start_h = (height - new_height) // 2
    start_w = (width - new_width) // 2
    cropped_zoomed_image = image[start_h:start_h+new_height, start_w:start_w+new_width]
    
    return cropped_zoomed_image

# Source directory containing the images
source_directory = r"C:\Users\Asus\OneDrive\Documents\School stuff\y 23-24\2nd sem\des2\Tuna_GRADES-20230522T082341Z-001\Tuna_GRADES\testing"  # Replace "path/to/source/images" with the actual path to your source image directory

# Target directory for resized images
target_directory = r"C:\Users\Asus\OneDrive\Documents\School stuff\y 23-24\2nd sem\des2\Tuna_GRADES-20230522T082341Z-001\Tuna_GRADES\restest"  # Replace "path/to/target/images" with the desired output directory for resized images

# Target size for resizing (square)
target_size = 800

# Zoom factors for the four zoomed-in copies
zoom_factors = [1.5, 2.0, 2.5, 3.0]

# Create the target directory if it doesn't exist
if not os.path.exists(target_directory):
    os.makedirs(target_directory)

# Iterate over images in the source directory
for filename in os.listdir(source_directory):
    if filename.endswith(".jpg") or filename.endswith(".jpeg"):  # Look for JPEG files
        # Load, crop, resize, and remove background from the image
        input_file_path = os.path.join(source_directory, filename)
        processed_image = load_crop_resize_remove_background(input_file_path, target_size)
        
        if processed_image is not None:
            # Save the original size processed image
            original_output_file_path = os.path.join(target_directory, os.path.splitext(filename)[0] + "_original.png")
            cv2.imwrite(original_output_file_path, processed_image)

            # Save the zoomed-in copies
            for i, zoom_factor in enumerate(zoom_factors):
                # Crop zoom the processed image
                cropped_zoomed_image = crop_zoom(processed_image, zoom_factor)
                
                # Add alpha channel
                cropped_zoomed_image = cv2.cvtColor(cropped_zoomed_image, cv2.COLOR_BGR2BGRA)
                
                # Set alpha channel to 0 for background pixels
                cropped_zoomed_image[..., 3] = np.where(cropped_zoomed_image[..., :3] == [0, 0, 0], 0, 255)
                
                # Save the zoomed-in image with transparent background
                zoomed_output_file_path = os.path.join(target_directory, os.path.splitext(filename)[0] + f"_zoom_{i+1}.png")
                cv2.imwrite(zoomed_output_file_path, cropped_zoomed_image)

print("All images processed and saved to:", target_directory)
```

```python
import cv2
import numpy as np
import os
from skimage.feature.texture import graycomatrix, graycoprops
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Function to convert RGB to HSI
def rgb_to_hsi(image):
    with np.errstate(divide='ignore', invalid='ignore'):
        r, g, b = cv2.split(image.astype(float) / 255)
        intensity = (r + g + b) / 3.0
        min_rgb = np.minimum(np.minimum(r, g), b)
        saturation = 1 - (3 / (r + g + b + 1e-10) * min_rgb)
        hue = np.zeros_like(r)
        num = 0.5 * ((r - g) + (r - b))
        den = np.sqrt((r - g) * (r - g) + (r - b) * (g - b))
        den[den == 0] = 1e-10
        theta = np.arccos(num / den)
        hue[b > g] = 2 * np.pi - theta[b > g]
        hue[b <= g] = theta[b <= g]
        hue = np.degrees(hue)
        return cv2.merge([hue, saturation, intensity])

# Function to extract features from a single image
def extract_features(image):
    if image.shape[-1] == 4:
        alpha_channel = image[:, :, 3]
        non_zero_pixels = np.where(alpha_channel != 0)

        if len(non_zero_pixels[0]) == 0:
            return None, None, None, None

        rgb = image[non_zero_pixels[0], non_zero_pixels[1], :3]
        
        hsv = cv2.cvtColor(rgb.reshape(-1, 1, 3), cv2.COLOR_RGB2HSV).reshape(-1, 3)
        hsi = rgb_to_hsi(rgb.reshape(-1, 1, 3)).reshape(-1, 3)

        gray = cv2.cvtColor(rgb, cv2.COLOR_RGB2GRAY)
        
        if gray.ndim == 2 and gray.shape[0] > 1 and gray.shape[1] > 1:
            glcm = graycomatrix(gray, distances=[1], angles=[0, np.pi/4, np.pi/2, 3*np.pi/4],
                                symmetric=True, normed=True)
            glcm_features = {prop: graycoprops(glcm, prop).ravel() for prop in ['contrast', 'correlation', 'energy', 'homogeneity']}
        else:
            glcm_features = {}

        return rgb, hsv, hsi, glcm_features

    else:
        rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        hsv = cv2.cvtColor(rgb, cv2.COLOR_RGB2HSV)
        hsi = rgb_to_hsi(rgb)

        gray = cv2.cvtColor(rgb, cv2.COLOR_RGB2GRAY)
        
        if gray.ndim == 2 and gray.shape[0] > 1 and gray.shape[1] > 1:
            glcm = graycomatrix(gray, distances=[1], angles=[0, np.pi/4, np.pi/2, 3*np.pi/4],
                                symmetric=True, normed=True)
            glcm_features = {prop: graycoprops(glcm, prop).ravel() for prop in ['contrast', 'correlation', 'energy', 'homogeneity']}
        else:
            glcm_features = {}

        return rgb, hsv, hsi, glcm_features

# Function to display prominent colors
def display_prominent_colors(image, n_colors=5):
    pixels = image.reshape(-1, 3)
    kmeans = KMeans(n_clusters=n_colors)
    kmeans.fit(pixels)
    colors = kmeans.cluster_centers_
    colors = colors.astype(int)
    palette = np.zeros((50, 300, 3), dtype='uint8')
    step = 300 // n_colors
    for i, color in enumerate(colors):
        palette[:, i * step:(i + 1) * step] = color

    plt.figure(figsize=(10, 2))
    plt.axis('off')
    plt.imshow(palette)
    plt.show()

# Directory containing the images
directory = r"C:\Users\Asus\OneDrive\Documents\School stuff\y 23-24\2nd sem\des2\Tuna_GRADES-20230522T082341Z-001\Tuna_GRADES\restest"

# List to store features for all images
all_features = []

# Iterate over images in the directory
for filename in os.listdir(directory):
    if filename.endswith(".jpg") or filename.endswith(".png"):
        image_path = os.path.join(directory, filename)
        image = cv2.imread(image_path, cv2.IMREAD_UNCHANGED)

        if image is not None:
            features = extract_features(image)
            if features:
                rgb, hsv, hsi, glcm = features
                all_features.append({
                    'filename': filename,
                    'rgb': rgb,
                    'hsv': hsv,
                    'hsi': hsi,
                    'glcm': glcm
                })
                display_prominent_colors(rgb)

print(all_features)

# Convert all_features list to a DataFrame
# df = pd.DataFrame(all_features)

# Save DataFrame to a CSV file
# df.to_csv('test_image_features.csv', index=False)
```

```python
import cv2
import numpy as np
from skimage.feature.texture import graycomatrix, graycoprops
import os

# Function to convert RGB to HSI
def rgb_to_hsi(image):
    r, g, b = image[:, :, 0], image[:, :, 1], image[:, :, 2]
    intensity = (r + g + b) / 3.0
    saturation = 1 - (3.0 / (r + g + b + 1e-10) * np.minimum(np.minimum(r, g), b))
    hue = np.arccos((0.5 * ((r - g) + (r - b))) / (np.sqrt((r - g)**2 + (r - b) * (g - b)) + 1e-10))
    hue[b > g] = 2 * np.pi - hue[b > g]
    hue = np.degrees(hue)
    return np.dstack((hue, saturation, intensity))

# Function to extract features from an image
def extract_features(image_path):
    # Load image
    image = cv2.imread(image_path)
    
    # Extract RGB features
    rgb_features = image
    
    # Convert to HSV
    hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    
    # Extract HSV features
    hsv_features = hsv_image
    
    # Convert to HSI
    hsi_image = rgb_to_hsi(image)
    
    # Extract HSI features
    hsi_features = hsi_image
    
    # Convert to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Compute GLCM properties
    glcm = graycomatrix(gray_image, distances=[1], angles=[0, np.pi/4, np.pi/2, 3*np.pi/4], levels=256, symmetric=True, normed=True)
    glcm_props = graycoprops(glcm, 'contrast'), graycoprops(glcm, 'dissimilarity'), graycoprops(glcm, 'homogeneity'), graycoprops(glcm, 'ASM'), graycoprops(glcm, 'energy')
    
    # Concatenate features
    features = np.concatenate((rgb_features.flatten(), hsv_features.flatten(), hsi_features.flatten(), np.array(glcm_props).flatten()))
    
    return features

# Directory containing images
image_dir = r"C:\Users\Asus\OneDrive\Documents\School stuff\y 23-24\2nd sem\des2\Tuna_GRADES-20230522T082341Z-001\Tuna_GRADES\restest"

# List to store features and corresponding labels
features = []
labels = []

# Iterate over images in the directory
for filename in os.listdir(image_dir):
    if filename.endswith(".jpg") or filename.endswith(".png"):
        # Extract grade from filename (assuming filename format: grade_filename.jpg)
        grade = filename.split('_')[0]
        
        # Extract features
        image_path = os.path.join(image_dir, filename)
        image_features = extract_features(image_path)
        
        # Append features and corresponding label
        features.append(image_features)
        labels.append(grade)

# Convert lists to arrays
features = np.array(features)
labels = np.array(labels)

# Now, features contains the extracted features and labels contains the corresponding grades.
# You can use these for KNN image processing.
```
